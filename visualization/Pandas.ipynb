## Identifying Missing Values

- Missing values are typically shown as NaN or None in Pandas
- alternatively, they might be:
    - zero
    - false
    - not applicable
    - entered an empty string
- they could: 
    - hint at data collection errors
    - indicate improper conversion or manipulation
    - not considered as missing
- there should not be missing values when graphing data

### Finding Missing Values

```python
df.info()
```

.info shows that there is one missing value in interest_paid column as there are only 59 non-null values

```python
df['interest_paid'].isna().head()
```

```python
interest_missing = df['interest_paid'].isna()
```

```python
df.loc[interest_missing, :]
```

```python
df.loc[~interest_missing, :]
```

" ~ " excludes the row with NaN value

```python
df["interest_paid"].isna().sum()
```

- This code counts the number of missing values (shown as "true" b/c the filter)
- .sum() works here because in python, booleans are a subtype of integers.

## Removing Missing Data

```python
df.info()
```

- In pandas, .dropna removes null values.
- It drops entire rows if they contain "any" or "all" nans--this may not be the best strategy depending on dataset

```python
df[30:40].dropna(how = "any")
```

## Filling in Missing Values

### .fillna

```python
df['interest_paid'][30:40]
```

```python
df['interest_paid'][30:40].fillna(0)
```

.fillna() fills the data with the set value (0.00 here)
- filling in zero may lead one to miss interpret the data

```python
df['interest_paid'][30:40].fillna(method='bfill')
```

'bfill' (back fill) fills NaN with the value after it
- more commonly done with time series data
- similarly, 'ffill' (foward fill) fills NaN with the value before it
    - choosing which one to use depending on context

### Linear Interpolation

```python
df['interest_paid'][30:40].interpolate(method = 'linear')
```

- linear interpolation uses a linear model to calculate the missing value

```python
interest_missing = df['interest_paid'].isna
```

```python
df.loc["interest_missing", 'interest_paid'] = 93.24
```

```python
df['interest_paid'][30:40].fillna(93.24)
```

```python
df['interest_paid'].sum()
```

^ this number is different from the original one before filling in the value
